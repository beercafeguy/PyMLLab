```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import ensemble
from sklearn import metrics
from sklearn import model_selection
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
X = df.drop("price_range",axis=1).values
y = df.price_range.values
```

## grid search Cross validation

```python
classifier = RandomForestClassifier(n_jobs=-1)
params_grid = {
    'criterion':['gini','entropy'],
    'max_depth':[1,3,5,7],
    'n_estimators':[100,200,300,400]
}

model = model_selection.GridSearchCV(
    estimator=classifier,
    param_grid=params_grid,
    scoring='accuracy',
    verbose = 10,
    n_jobs=1,
    cv = 5
)
model.fit(X,y)
```

```python
print(model.best_score_)
print(model.best_estimator_.get_params())
print(model.best_params_) # only what I passed
```

```python
from sklearn import decomposition
from sklearn import preprocessing
from sklearn import pipeline
```

```python
## Param tuning with Pipeline
```

```python
std_scl = preprocessing.StandardScaler()
pca = decomposition.PCA()
rf_classifier = RandomForestClassifier(n_jobs=-1)


clf_pipeline = pipeline.Pipeline(
    [
        ('scaling',std_scl),
        ('pca',pca),
        ('model',rf_classifier)
    ]
)


params_grid = {
    'pca__n_components':np.arange(5,10),
    'model__criterion':['gini','entropy'],
    'model__max_depth':[1,3,5,7],
    'model__n_estimators':[100,200,300,400]
}

model = model_selection.RandomizedSearchCV(
    estimator=clf_pipeline,
    param_distributions=params_grid,
    n_iter=10,
    scoring='accuracy',
    verbose = 10,
    n_jobs=1,
    cv = 5
)
model.fit(X,y)
```

## bayesian optimization

```python
from ast import Param
def optimize(params_values,param_names,x,y):
  params = dict(zip(param_names,params_values))
  model = RandomForestClassifier(**params)

  kf = model_selection.StratifiedKFold(n_splits=5)

  accuracies = []
  for idx in kf.split(X = x,y=y):
    train_idx,test_idx = idx[0],idx[1]

    x_train = x[train_idx]
    y_train = y[train_idx]

    x_test = x[test_idx]
    y_test = y[test_idx]

    model.fit(x_train,y_train)

    y_pred = model.predict(x_test)
    fold_acc = metrics.accuracy_score(y_test,y_pred)

    accuracies.append(fold_acc)

  return -1 * np.mean(accuracies) # because we need to maximize the accuracy



# define search space

from skopt import space
param_space = [
    space.Integer(3,15,name='max_depth'),
    space.Integer(100,600,name='n_estimators'),
    space.Real(0.01,1.0,prior="uniform",name="max_features"),
    space.Categorical(["gini","entropy"],name="criterion")
]

param_names = ["max_depth","n_estimators","max_features","criterion"]

from functools import partial
optimization_function = partial(
    optimize,
    param_names = param_names,
    x= X,
    y = y
)

from skopt import gp_minimize
result = gp_minimize(
    optimization_function,
    dimensions=param_space,
    n_calls = 15,
    n_random_starts=10,
    verbose=10
)
```

### Hyperopt

```python
from hyperopt import hp,fmin,tpe,Trials,STATUS_OK
from hyperopt.pyll.base import scope
```

```python
from ast import Param
def optimize(params,x,y):
  model = RandomForestClassifier(**params)

  kf = model_selection.StratifiedKFold(n_splits=5)

  accuracies = []
  for idx in kf.split(X = x,y=y):
    train_idx,test_idx = idx[0],idx[1]

    x_train = x[train_idx]
    y_train = y[train_idx]

    x_test = x[test_idx]
    y_test = y[test_idx]

    model.fit(x_train,y_train)

    y_pred = model.predict(x_test)
    fold_acc = metrics.accuracy_score(y_test,y_pred)

    accuracies.append(fold_acc)

  return -1 * np.mean(accuracies) # because we need to maximize the accuracy
```

```python
param_space = {
    'max_depth': scope.int(hp.quniform('max_depth',3,15,1)),
    'max_depth': scope.int(hp.quniform('n_estimators',100,600,1)),
    'max_features': hp.uniform('max_features',0.01,1.0),
    'criterion': hp.choice('criterion',["gini","entropy","log_loss"])
}
```

```python
trials = Trials()
optimization_function = partial(optimize,x=X,y = y)
result = fmin(
    optimization_function,
    space = param_space,
    trials = trials,
    max_evals = 15,
    algo=tpe.suggest
)

print(result)
```

```python
import optuna

def optimize(trial,x,y):

  criterion = trial.suggest_categorical("criterion",["gini","entropy"])
  n_estimators = trial.suggest_int("n_estimators",100,1500)
  max_depth = trial.suggest_int("max_depth",3,15)
  max_features = trial.suggest_float("max_features",0.01,1.0)

  model = RandomForestClassifier(
      n_estimators=n_estimators,
      max_depth = max_depth,
      max_features = max_features,
      criterion = criterion
  )

  kf = model_selection.StratifiedKFold(n_splits=5)

  accuracies = []
  for idx in kf.split(X = x,y=y):
    train_idx,test_idx = idx[0],idx[1]

    x_train = x[train_idx]
    y_train = y[train_idx]

    x_test = x[test_idx]
    y_test = y[test_idx]

    model.fit(x_train,y_train)

    y_pred = model.predict(x_test)
    fold_acc = metrics.accuracy_score(y_test,y_pred)

    accuracies.append(fold_acc)

  return -1 * np.mean(accuracies) # because we need to maximize the accuracy


study = optuna.create_study(direction="minimize")
optimization_function = partial(optimize,x=X,y = y)
study.optimize(
    optimization_function,
    n_trials=15
)
```
