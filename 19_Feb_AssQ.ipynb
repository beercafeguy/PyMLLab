**Q3.** What is App routing in Flask? Why do we use app routes?

**Ans:**

App routing in flask is a way to bind the perticular URL with a perticular function call. 
<br>For example, if I want to load my user_details page when I call '/user' in url so I will use app routing for the same. 
<br>
URL routing is also used to bind different functions with different URLs in API programming.
<br> For Example I get user '/users?id=<id>' to get the details of a user where as I can use  '/courses?course_id=<id>' to get the details of course. There will be two different functions mapped to these URLs which will handle these two requests.

**Q5.** What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

**Ans**

redirect in combination with url_for() are used for URL building in flask.

```python
from flask import Flask,url_for,redirect
from flask import request

app = Flask(__name__)

data  = {
    "hem":"Developer",
    "ajay":"QA",
    "sahil":"HR"
}


@app.route("/")
def home():
    return "<h1>Welcome to ABC Limited!</h1>"


@app.route("/dev")
def dev():
    return "<h1>I am a developer</h1>"


@app.route(("/qa"))
def qa():
    return "<h1>I am a QA</h1>"

@app.route("/hr")
def hr():
    return "<h1>I am HR</h1>"

@app.route(("/user"))
def get_name():
    name = request.args.get("name")
    print("Entered name {}".format(name))
    if name in data.keys():
        role = data[name]
        if role == "Developer":
            return redirect(url_for('dev'))
        elif role == "QA":
            return redirect(url_for('qa'))
        elif role == "HR":
            return redirect(url_for('hr'))
        else:
            return "Invalid User {}".format(name)
    else:
        return "Invalid User {}".format(name)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

In above codebase, we have three different functions mapped to three different endpoints (dev/qa/hr). Here using '/user' routing, we are getting the name of the user and redirecting the page to a url based on the ROLE of the user. (Roles are predefind in codebase.)

If we geta username which is not available in mapping, we redirect that to en error message  "Invalid User."
