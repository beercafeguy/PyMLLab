**Q4.** What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

**Ans:**

DataFrame is a two dimentional labled tabular structure. This is madeup of one or more then one series with common indices. Dataframe is two dimentional where as Series is a one dimentional structure. Dataframe contains columns with column names where as Series is just a key value pair list kind of structure where key is indice of the series.

```python
# Series

sr = pd.Series([1,2,3,4,5])
print(sr)

# here this is a series of 5 numbers.
# the first column below prints the indices
```

**Q5.** What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

**Common Functions used in pandas Dataframe**

1. isna()
2. value_counts()
3. group_by()
4. loc() and iloc()
5. unique() and nunique()
6. head() and tail()
7. info()
8. describe()
9. drop_duplicates()
10. sort_values()

```python
# example of loc and iloc

# loc -> Location
# iloc -> indexed Location
# lets prepare sample data 
td = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
td.head()
```

```python
# loc is used for value based selection like below

td.loc[0:2,["PassengerId","Survived"]]
```

```python
# loc is used for index based selection like below
td.iloc[0:3,0:2]
```

```python
td.loc[(td.Pclass == 3) & (td.Sex == "male")]
```
