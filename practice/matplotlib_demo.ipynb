```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
y[0:10]
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
colours = np.random.rand(50)
plt.scatter(x,y,c=colours)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
sizes = 100 * (np.random.rand(50))
colours = np.random.rand(50)
plt.scatter(x,y,s=sizes,c=colours)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
sizes = 100 * (np.random.rand(50))
colours = np.random.rand(50)
plt.scatter(x,y,s=sizes,c=colours,alpha=0.5)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```
