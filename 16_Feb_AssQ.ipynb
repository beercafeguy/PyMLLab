**Q1.** What is a database? Differentiate between SQL and NoSQL databases.

**ANS**

**Database **: Database is a software system which is used to store the data in an organized fashion on computer disks. It provides different constructs for managing the data, options for access management and reliable ways of managing data.

**Difference between SQL and NoSQL databases**

SQL databases are the systems where data can be stores and accessed using a standard query language called SQL. All the SQL databases supports some form of SQL for there operations. Example : MySQL, SQL Server, Oracle 11G, DB2 etc

<br>

NoSQL (Not only SQL) databases are databses are designed for specific neeeds where RDBMS systems might not be best fit like specific scale and latency needs. Mostly these systems do not support SQL in plain form. Each NoSQL databases has its own API/language for managing and accessing data. Example of such databases are MongoDB, Cassandra etc.

**Q2.** What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

**DDL:**

Data Definition language (DDL) are the set of SQL commands which are used for creating and managing RDBMS storage constructs like databases and tables.

<br>CREATE: Used to create the object (table/database)
<br>DROP: Used to drop the object from database (delete full structure and data)
<br>ALTER: Change the structure of the object like adding columns in table.
<br>TRUNCATE: Delete all the data from table but keep the structure.
