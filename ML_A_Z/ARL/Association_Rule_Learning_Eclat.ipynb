```python
path = '/content/drive/MyDrive/ML_A_Z/Part_5_Association_Rule_Learning/Section_29_Eclat/Python/Market_Basket_Optimisation.csv'
df = pd.read_csv(path,header=None)
df.head()
```

```python
transactions = []
for i in range(0,df.shape[0]):
  transactions.append([str(df.values[i][j]) for j in range(0,df.shape[1])])
transactions[0]
```

```python
from apyori import apriori

rules = apriori(
    transactions=transactions,
    min_support = 0.003, # 3 items per day for a week / total transactions
    min_confidence = 0.2,
    min_lift = 3,
    min_length=2,
    max_length = 2
)
```

### Visualize the result

```python
results = list(rules)
results
```

```python
len(results)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product1', 'Product2', 'Support'])
```

```python
resultsinDataFrame.sort_values(by='Support',ascending=False)
```

```python
resultsinDataFrame.nlargest(n=10,columns=['Support'])
```
