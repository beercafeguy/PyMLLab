```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
y[0:10]
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
colours = np.random.rand(50)
plt.scatter(x,y,c=colours)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
sizes = 100 * (np.random.rand(50))
colours = np.random.rand(50)
plt.scatter(x,y,s=sizes,c=colours)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
sizes = 100 * (np.random.rand(50))
colours = np.random.rand(50)
plt.scatter(x,y,s=sizes,c=colours,alpha=0.5)
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)

plt.bar(x,y)
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)

plt.barh(x,y)
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)

plt.figure(figsize=(12,8))
plt.plot(x,y,'--b')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,c='r',s=100*np.random.rand(50),marker='x')
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,c='r',s=100*np.random.rand(50),marker='v')
plt.xlabel('random number')
plt.ylabel('random number')
plt.show()
```

```python
x = [4,5,6,7,8,9]
y = [6,7,8,9,10,11]
color = ['r','b','g','#D05034','#4B8151','#D43AD4']
plt.scatter(x,y,c=color)
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
