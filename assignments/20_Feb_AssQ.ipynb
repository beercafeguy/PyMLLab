**Q1.** Explain GET and POST methods.

**Ans:**

**GET**
Get request is used to send the data to server through request URLs. This can be bookmarked and cached in browser history. These kind of request are mainly used for requesting some kind of data from web server based on the input values passed in URL.

**POST**
POST method is also used to end the data to server but to create or update a resource. Here the data goes as part of request body rather than request URL. So it makes it safe compare to GET request. These request can't be cached and can't be bookmarked. Here in case of POST we don't have any restriction on data unlike GET where we have certain restrictions. But never the less, we need to be mindfull of the data volume which we are passing over network.

**Q4.** What are templates in Flask? Why is the render_template() function used?

**And:**

**Templates** A template is a file which can contain fixed as well as dynamic content for a visualization later. In standard HTML page, we can put only static content but in templates we can put if else statements, variables, loops etc. This is used to load the HTML page with dynamic content easily.

Flask support Jinja tempelating framework for creating HTML templates.

**Use of render_template function**

render_template function in Flask used used to send to response for a request as a template. This also allows to pass variables from server side to template so that same can be used to load the final HTML page.
