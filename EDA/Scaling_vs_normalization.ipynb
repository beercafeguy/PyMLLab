```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.preprocessing import minmax_scaling
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
sample_data = np.random.exponential(size=2000)
sample_data_min_max_scaled = minmax_scaling(sample_data,columns=0)
sample_data_min_max_scaled
```

```python
min_max_scaler = MinMaxScaler()
sample_data_reshaped = sample_data.reshape(len(sample_data),1)
min_max_scaled = min_max_scaler.fit_transform(sample_data_reshaped).reshape(1,len(sample_data))
min_max_scaled[0]
```

```python
std_scaler = StandardScaler()
std_scaled = std_scaler.fit_transform(sample_data_reshaped).reshape(1,len(sample_data))[0]
std_scaled
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,8))
sns.histplot(sample_data, ax=ax[0][0], kde=True, legend=False)
ax[0][0].set_title("Original Data")
sns.histplot(sample_data_min_max_scaled, ax=ax[0][1], kde=True, legend=False)
ax[0][1].set_title("Min Max Scaled Data")
sns.histplot(min_max_scaled[0], ax=ax[1][0], kde=True, legend=False)
ax[1][0].set_title("Min Max Scaled Data SKLEARN")
sns.histplot(std_scaled, ax=ax[1][1], kde=True, legend=False)
ax[1][1].set_title("Standard Scaled Data SKLEARN")
```
