```python
path = '/content/drive/MyDrive/ML_A_Z/Part_4_Clustering/Section_25_Hierarchical_Clustering/Python/Mall_Customers.csv'
df = pd.read_csv(path)
df.head()
```

```python
X = df.iloc[:,[3,4]].values # customerId is not a good feature as it does not hold any information
```

```python
%matplotlib inline

plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],color='red')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Income data scatter plot')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage

dend = dendrogram(linkage(X,method='ward'))
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.title('dendrogram')
plt.show()
```

```python
#optimal number of clusters is 5
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
```

```python
#other way of finding predictions
y_clustering = clustering.fit_predict(X)
```

```python
y_clustering
```

```python
%matplotlib inline

plt.scatter(X[y_clustering == 0,0],X[y_clustering == 0,1],color='red',label='cluster 1')
plt.scatter(X[y_clustering == 1,0],X[y_clustering == 1,1],color='blue',label='cluster 2')
plt.scatter(X[y_clustering == 2,0],X[y_clustering == 2,1],color='green',label='cluster 3')
plt.scatter(X[y_clustering == 3,0],X[y_clustering == 3,1],color='orange',label='cluster 4')
plt.scatter(X[y_clustering == 4,0],X[y_clustering == 4,1],color='yellow',label='cluster 5')

#plt.scatter(clustering.cluster_centers_[:,0],clustering.cluster_centers_[:,1],color='magenta',s=300,label='centroid')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Plotted clusters')
plt.legend()
plt.show()
```

```python
# for 3 clusters

clustering = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_clustering = clustering.fit_predict(X)
```

```python
%matplotlib inline

plt.scatter(X[y_clustering == 0,0],X[y_clustering == 0,1],color='red',label='cluster 1')
plt.scatter(X[y_clustering == 1,0],X[y_clustering == 1,1],color='blue',label='cluster 2')
plt.scatter(X[y_clustering == 2,0],X[y_clustering == 2,1],color='green',label='cluster 3')

#plt.scatter(clustering.cluster_centers_[:,0],clustering.cluster_centers_[:,1],color='magenta',s=300,label='centroid')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Plotted clusters')
plt.legend()
plt.show()
```
